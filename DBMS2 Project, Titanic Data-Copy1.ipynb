```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
data.columns = ('PassengerId', 'Survived', 'Pclass', 'Name',
               'Sex', 'Age', 'Sibsp', 'Parch', 'Ticket', 'Fare',
               'Cabin', 'Embarked')

#dropping 2 columns that aren't useful atm
data = data.drop(["PassengerId"], axis=1)
data = data.drop(["Cabin"], axis=1)

print("Number of instances = %d" % (data.shape[0]))
print("Number of attributes = %d" % (data.shape[1]))

data.head(50)
```

```python
print("Number of instances = %d" % (data.shape[0]))
print("Number of attributes = %d" % (data.shape[1]))

print("Number of missing values: ")
for i in data.columns:
    print('\t%s: %d' % (i, data[i].isna().sum()))

data.Embarked.fillna(data.Embarked.mode()[0], inplace = True)
```

```python
data2 = data['Age']
# Looking at 50 instances to see the frequency of NaN values in Age column
print('Before modification:')
print(data2[42:50])
# repalcing the NaN values with the mean of that column, applying the round function
data2 = data2.fillna(round(data2.mean()))

print('\nAfter modification:')
print(data2[42:50])
```
